```python

import yaml, pandas as pd

with open('../configs/data.yml') as f:
  dt = yaml.safe_load(f)

path_dt = f"../{dt['mpdata']}"
data = pd.read_csv(path_dt)


yname = "lemp"
gname = "first.treat"
idname = "countyreal"
tname = "year"
xformla = f"lemp~1"
```

```python
b = ATTgt(yname, tname, idname, gname, data = data, xformla=xformla).fit()
b.plot_attgt()
```
